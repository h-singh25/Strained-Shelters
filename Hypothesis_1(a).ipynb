So far we have cleaned the column names in ```income_data```. The next step is to handle the missing values in the data. 

There are two possibilities to fill the missing values in data, using mean or median. To make the decision out of these two which one to be used we performed some plotting below.

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean style
sns.set(style="whitegrid")

# selecting numeric columns
numeric_cols = income_data.select_dtypes(include='number').columns

# Choose a color palette with multiple colors
palette = sns.color_palette("husl", len(numeric_cols))

# Plot histogram for each numeric column for skewness
for i, col in enumerate(numeric_cols):
    plt.figure(figsize=(4, 2))
    sns.histplot(income_data[col], kde=True, bins=15, color=palette[i])
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
````

````python
# boxplot for checking the outliers
for col in numeric_cols:
    plt.figure(figsize=(4, 2))
    sns.boxplot(x=income_data[col])
    plt.title(f'Boxplot of {col}')
    plt.tight_layout()
    plt.show()
````

To replace the missing values in ```income_data```, there are two possibilities : replaces NaNs with ```mean``` or replace NaNs with ```median```. <br> The idea is to group all the household types according to the year using ```groupby```-method and then replace the NaNs in all columns (having numeric values only) with either the ```mean``` or ```median``` for that group. <br> 

For this task we check for the distribution of the data to check for the skewness in the data. And also the box-plots to check the presence of outliers. 

##### Histograms :<br>
- We used Histograms to check the distribution of the data. We see that for all the numeric columns the data is usually right-skewed which makes the choice of ```median``` better for the replacement. Because the mean would usually shift towards the tail as the skewness shifts.


##### Box-plots :<br>
- Similary, in box-plots we see the outliers along with the possible mean shift in the data. We can see long tails or outliers, choosing ```mean``` in this case can be misleading because it gets pulled in the direction of the extreme values. To make our data more robust to outliers and skewness, ```median``` would be the better choice.  

From our findings we came on the conclusion to replace the missing values with ```median``` values of each group to avoid shifts in data-mean and for more robust base for further analysis. 

````python
# fill NaNs dynamically for all columns 
# selecting only the columns with numeric values
numeric_columns = income_data.select_dtypes(include='number').columns

# Grouping by year and fill NaNs with the mean of each group
income_data[numeric_columns] = income_data.groupby('year')[numeric_columns].transform(
                                                                        lambda x: x.fillna(x.median())
                                                                        )  
````

### <u>Hypothesis 1(a)</u> :
_**People who live alone and Single Parents are at greatest risk of Poverty**_.<br> Single-person households are at higher risk of housing cost overburden
- Use % of housing cost poverty risk population and household types.
- Check trends : are Persons living alone consistently overburdened?
- Data to Use:
    - % poverty risk per year
    - Compare with Total population
    - Household breakdowns (especially single-person vs. two adults)"

````python
column_list
````

````python
# Filter only relevant household types
relevant_types = ['Persons living alone', 'Lone parents',
                  'Two adults without children', 'Two adults with children',
                  'Three or more adults with children'] 
subset_data = income_data[income_data['household_type'].isin(relevant_types)]
print(f"shape of subset_data  ::  \n{subset_data.shape}")

subset_data 
````

````python
import matplotlib.cm as cm
import matplotlib.colors as mcolors


# grouping to plot
grouped = subset_data.groupby('household_type')['poverty_risk_population'].mean().sort_values(
                                                                                        ascending=False)
# Reset Index to make a column again out of the index
grouped = grouped.reset_index()


# Normalize the data values to 0-1 range for colormap
norm = mcolors.Normalize(vmin=grouped['poverty_risk_population'].min(), 
                         vmax=grouped['poverty_risk_population'].max())
# Choose a colormap
cmap = cm.get_cmap('coolwarm')  # You can try 'coolwarm', 'plasma', 'magma', etc.
# Generate colors based on values
colors = [cmap(norm(value)) for value in grouped['poverty_risk_population']]



# Barplot
plt.figure(figsize=(8, 4))
ax = sns.barplot(data=grouped, 
                 x='poverty_risk_population', 
                 y='household_type', 
                 palette=colors)  # Try 'pastel', 'deep', etc.

# Adjusting bar thickness
for patch in ax.patches:
    patch.set_height(0.5)  # type: ignore # Adjust between 0.3 and 0.8 as needed

# Add labels
for patch in ax.patches:
    value = patch.get_width()
    y = patch.get_y() + patch.get_height() / 2
    ax.text(value + 0.5, y, f"{value:.1f}%", va='center', fontsize=9)

plt.title('Average Poverty Risk by Household Type')
plt.xlabel('Average Poverty Risk (%)')
plt.ylabel('Household Type')
plt.tight_layout()
plt.show()
````

- Average Poverty Risk over time by Household type. <br> We use ```Line-Plot``` to showcase this for the years 2020, 2021, 2022, 2023 and 2024.

````python
# Plot with a nice palette
plt.figure(figsize=(8, 4))
sns.lineplot(
    data=subset_data,
    x='year',
    y='poverty_risk_population',
    hue='household_type',
    marker='o',
    palette='Dark2'  
)
plt.title("Average Poverty Risk Over Time by Household Type")
plt.ylabel("Poverty Risk Population (%)")
plt.xlabel("Year")
plt.legend(title="Household Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
````

_**Interpretation :**_ The graph shows that people living alone and single parents consistently bear the highest risk of poverty in Germany, with peaks of over 50%. In 2021 in particular, the risk was extremely high for these groups, while multi-person households with children are significantly less affected. The trend thus clearly confirms that single-person households and single parents are particularly at risk and are above the social average.

- We can also see in the plot below more detailed visualization for this using bar plots for different years.

````python
subset_data.household_type.unique()
````

````python
# Adjustment for the main plot
years = sorted(subset_data['year'].unique())
for yr in years:
    year_subset = subset_data[subset_data['year'] == yr]

n_cols = 2
n_rows = (len(years) + n_cols - 1) // n_cols  # calculate needed rows


custom_colors = ["#EBA817", "#56B1E5", "#10B78A", "#C92EAA", "#589BC2"]
# Repeat colors if there are more bars than colors:
colors = [custom_colors[i % len(custom_colors)] for i in range(len(year_subset))]
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 4), constrained_layout=False)


# Flatten axes array for easy iteration (in case n_rows or n_cols = 1)
axes = axes.flatten()

for i, yr in enumerate(years):
    ax = axes[i]
    year_subset = subset_data[subset_data['year'] == yr]
    
    # Sort if you want bars ordered by value
    year_subset = year_subset.sort_values('poverty_risk_population')
    
    ax.barh(year_subset['household_type'], year_subset['poverty_risk_population'], 
            color=colors)
    
    ax.set_title(f"Poverty Risk ({yr})")
    ax.set_xlabel("Poverty Risk Population (%)")
    ax.set_ylabel("Household Type")
    

    # Optional: improve layout for long labels
    ax.tick_params(axis='y', labelrotation=0)
    # Add value labels
    for index, value in enumerate(year_subset['poverty_risk_population']):
        ax.text(value + 0.5,   # shift right of bar end
                index,         # y position (matches bar)
                f"{value:.1f}%",  # formatted label
                va='center',
                fontsize=9)

# Hide unused subplots if any
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle("Poverty Risk by Household Type Over Years", fontsize=16)
plt.subplots_adjust(wspace=0.7, hspace=0.7, top=0.9)

plt.show()

````

For the year 2024, we see the results are a bit different where ```'three or more adults with children'``` have surpassed the ```'Single parents'```. However, this inconsistency could be due to the absence of enough data. Since we have only the data for the first quarter of the year 2024.
